```python
import geopandas as gpd
from shapely.geometry import mapping
import pandas as pd
from shapely.geometry import Point
```

```python
gdf_polygons = gpd.read_file('data/World_EEZ_v11_20191118/eez_v11.shp')
gdf_polygons
```

```python
df = pd.read_csv("data/lists-Reefers-2022-11-11_04-40.csv")
# df = df[['MMSI', 'Latitude', 'Longitude']]
# df = df.dropna()
```

```python
df['coordinates'] = list(zip(df['Longitude'], df['Latitude']))
df['coordinates'] 
```

```python
df.coordinates = df.coordinates.apply(Point)
df.coordinates
```

```python
gdf_points = gpd.GeoDataFrame(df, geometry='coordinates')
gdf_points
```

```python
gdf_points.crs = gdf_polygons.crs
gdf_points.crs
```

```python
gdf_polygons
```

```python
sjoin = gpd.sjoin(gdf_points, gdf_polygons, how='left', predicate='within')
```

```python
df_join = pd.DataFrame(sjoin)
sjoin
```

```python
df_join
```

```python
result = df_join #[["MMSI", "Latitude", "Longitude", "MRGID_TER1", "TERRITORY1", "TERRITORY2","SOVEREIGN1", "SOVEREIGN2"]]
```
